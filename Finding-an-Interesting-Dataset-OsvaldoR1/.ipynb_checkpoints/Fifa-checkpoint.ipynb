```python
import pandas
import seaborn as sns

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

fifa = pandas.read_csv("FIFA22_official_data.csv")
fifa.head(5)
```

```python
fifa = fifa.sort_values(by='Wage', ascending=False)

# Select the top 5 rows of the sorted DataFrame
top_5_wages = fifa.nlargest(5, 'Wage')

# Sort the DataFrame by 'Overall Rating' in descending order
fifa = fifa.sort_values(by='Best Overall Rating', ascending=False)

# Select the top 5 rows of the sorted DataFrame
top_5_overall = fifa.head(5)

# Create a list of the top 5 overall ratings
top_5_overall_ratings = top_5_overall['Best Overall Rating'].tolist()

# Check if the top 5 overall ratings are in the top 5 wages
result = all(rating in top_5_waged['Best Overall Rating'].tolist() for rating in top_5_overall_ratings)

# Print the result
print(result)
```
