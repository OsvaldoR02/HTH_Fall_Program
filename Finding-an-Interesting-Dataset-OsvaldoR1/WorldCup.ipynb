```python
import pandas

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

WD = pandas.read_csv("international_matches.csv")
WD.head(5)
```

```python
offense = WD[['date', 'home_team', 'home_team_mean_offense_score', 'away_team', 'away_team_mean_offense_score']]
home = offense[['date', 'home_team', 'home_team_mean_offense_score']]
away = offense[['date', 'away_team', 'away_team_mean_offense_score']]
ha = WD.groupby(['home_team', 'home_team_mean_offense_score', 'away_team', 'away_team_mean_offense_score'])
ha.nlargest(5,'home_team_mean_offense_score').sort_values(['date'], ascending=False)


```
